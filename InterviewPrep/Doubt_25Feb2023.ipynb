1 D Array 2D Array

```python
#Array 

a=['a','b','c','d']
print(a[0]) #--> Accessing by index
for i in a: #--> Accessing each value one by one
    print(i)
    
a.append('e') #--> Adding new item to array
print(a)
a.remove('c') #--> Remove  value in array 
print(a)
```

```python
#a=['a','b','c','d'] 1D

TwoDim = [['a','b'],['c','d']]

print(TwoDim[0]) #--> Accesing entire 0th index of 2D
print(TwoDim[1])

print(TwoDim[0][1]) #--> Accessing 0th index of 1st value in 2D array
```
