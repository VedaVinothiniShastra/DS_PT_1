Customizing Plot Legends	
    Choosing Elements for the Legend
	Legend for Size of Points
	Multiple Legends
	Customizing Colorbars


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-', label='Sine')
ax.plot(x, np.cos(x), '--', label='Cosine')
leg = ax.legend()
ax.legend(loc='upper right', frameon=True , ncol=2, borderpad=1, fancybox=True, shadow=True)
```

```python
#Multiple legend

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
lines=[]
styles=['-', '--', '-.', ':']
for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='black')
leg = ax.legend()
ax.legend(lines[:2], ['line A', 'line B'],loc='upper right', frameon=True)

from matplotlib.legend import Legend
l = Legend(ax, lines[2:], ['line C', 'line D'],loc='lower right', frameon=True)
ax.add_artist(l)
```

```python
#Customizing Colorbars

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
I =np.sin(x)*np.cos(x[:,np.newaxis])
plt.imshow(I, cmap='RdBu') #blues, gray 
plt.colorbar()
# cmpa='Binary' ->Mini Project
```

Multiple Subplots:
=> Append more images in single file 
    plt.axes: Subplots by Hand -> (left,bottom , width , height)
	plt.subplot: Simple Grids of Subplots
	plt.subplots: The Whole Grid 
	plt.GridSpec: More Complicated Arrangements

```python
#plt.axes
import matplotlib.pyplot as plt
import numpy as np
#a1=plt.axes() 
#a2=plt.axes([0.65,0.65,0.2,0.2]) #--> syntax

#add_axes()

f=plt.figure()
a1=f.add_axes([0.1,0.5,0.8,0.4])
a2=f.add_axes([0.1,0.1,0.8,0.4])
x=np.linspace(0,10,100)
a1.plot(np.sin(x))
a2.plot(np.cos(x))
```

```python
#plt.subplot(row, column , panel num)

for i in range(1,8): #-> 7 sub plots (1,8)
    plt.subplot(3,3,i)
    plt.text(0.5,0.5,str((3,3,i)),fontsize=20)
```

```python
for i in range(1,24): #-> 23 sub plots (1,24)
    plt.subplot(10,10,i)
    plt.text(0.5,0.5,str((10,10,i)),fontsize=5)
```

```python
#Adding space between the subplot
f=plt.figure()
f.subplots_adjust(hspace=0.4,wspace=0.4)
for i in range(1,24): #-> 23 sub plots (1,24)
    a=f.add_subplot(10,10,i)
    a.text(0.5,0.5,str((10,10,i)),fontsize=5)
```

```python
#plt.subplots

f,a=plt.subplots(4,5, sharex='col',sharey='row')
for i in range(4):
    for j in range(5):
        a[i,j].text(0.5,0.5,str((i,j)),fontsize=8)
```

```python
#plt.GridSpec  -> plt.subplot

grid=plt.GridSpec(2,3,wspace=0.4,hspace=0.3)
plt.subplot(grid[0,0]) #-> grid(Row, col) start from 0
plt.subplot(grid[0,1])
plt.subplot(grid[0,2])
plt.subplot(grid[1,0])
plt.subplot(grid[1,1])
plt.subplot(grid[1,2])

```

Text and Annotation	(subplots)
    Transforms and Text Position ->axes.transData, axes.TransAxes,f.transfigure
	Arrows and Annotation ->plt.arrow() , plt.annotate()

```python
#Transforms and Text Position
f,a=plt.subplots()
a.axis([0,10,0,10])
a.text(2,6,"(2,6)",transform=a.transData)
a.text(0.5,0.7,"(0.5,0.7)",transform=a.transAxes)
a.text(0.2,0.2,"(0.2,0.2)",transform=f.transFigure)
```

```python
#Arrows and Annotation ->plt.arrow() , plt.annotate()

f,a= plt.subplots()
x=np.linspace(0,20,100)
a.plot(x,np.cos(x))
a.axis('equal')
a.annotate('Local Max', xy=(5.7,1.5) ,xytext=(10,4) ,arrowprops=dict())
a.annotate('Local Min', xy=(9.5,-1.5), xytext=(2,-6),arrowprops=dict(arrowstyle='->'))
```

Customizing Ticks	
    Major and Minor Ticks
	Hiding Ticks or Labels
	Reducing or Increasing the Number of Ticks
	Fancy Tick Formats

```python
import matplotlib.pyplot as plt
import numpy as np
a=plt.axes(xscale='log',yscale='log')
a.grid()
#Hiding Ticks or Labels
a.yaxis.set_major_locator(plt.NullLocator())
a.xaxis.set_major_formatter(plt.NullFormatter()) 
```

```python
#Reducing or Increasing the Number of Ticks -> maclocator 
f, a= plt.subplots(2,2,sharex=True, sharey=True)
for i in a.flat:
    i.xaxis.set_major_locator(plt.MaxNLocator(2))
    i.yaxis.set_major_locator(plt.MaxNLocator(2))
```

```python
#Fancy Tick Formats ->MultipleLocator

f, a= plt.subplots(2,2,sharex=True, sharey=True)
for i in a.flat:
    i.xaxis.set_major_locator(plt.MaxNLocator(50/2)) #-> np.pi /2
    i.yaxis.set_major_locator(plt.MaxNLocator(50/2))
    
```
