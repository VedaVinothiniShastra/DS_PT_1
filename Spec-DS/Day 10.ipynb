High-Performance Pandas	
    Motivating query() and eval()
	Compound Expressions
	pandas.eval() for Efficient Operations
	DataFrame.eval() for Column-Wise Operations
	DataFrame.query() Method

```python
import numpy as np
rng=np.random.RandomState(42)
x=rng.rand(1000000)
y=rng.rand(1000000)
print(x+y)
%timeit x+y
```

```python
#eval()

import pandas as pd
row=100000
col=100
rang=np.random.RandomState(42)
df1, df2, df3, df4=(pd.DataFrame(rang.rand(row, col)) for i in range(4))
#print(df1)
#print(df2)
#print(df3)
#print(df4)

print(df1+df2+df3+df4)

%timeit df1+df2+df3+df4
```

```python
#pandas.eval() for Efficient Operations
%timeit pd.eval('df1+df2+df3+df4')
```

```python
#DataFrame.eval() for Column-Wise Operations
rang=np.random.RandomState(42)
a=pd.DataFrame(rang.rand(1000,3),columns=['A','B','C'])
print(a)
a.eval('D=(A+B+C)', inplace=True)
print(a)
```

```python
#Query
rang=np.random.RandomState(42)
a=pd.DataFrame(rang.rand(1000,3),columns=['A','B','C'])
print(a)
q=a.query('A<0.5 and B<0.5 and C<0.5')
print(q)
```

```python
import matplotlib.pyplot as plt
import numpy as np
X= np.linspace(0,10,100)
plt.figure() #->Create a fig in mat
Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = np.tan(X)
Y4 = np.tanh(X)
figure, axis = plt.subplots(2,2)
axis[0, 0].plot(X, Y1)
axis[0, 0].set_title("Sin")
axis[0, 1].plot(X, Y2)
axis[0, 1].set_title("Cos")
axis[1, 0].plot(X, Y3)
axis[1, 0].set_title("Tan")
axis[1, 1].plot(X, Y4)
axis[1, 1].set_title("Tanh")
plt.show()
```

```python
#Method 2 of 2 rows 2cols

import matplotlib.pyplot as plt
import numpy as np
X= np.linspace(0,10,100)
f=plt.figure() #->Create a fig in mat
Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = np.tan(X)
Y4 = np.tanh(X)
f.add_subplot(2,2,1)
plt.plot(X, Y1)
f.add_subplot(2,2,2)
plt.plot(X, Y2)
f.add_subplot(2,2,3)
plt.plot(X, Y3)
f.add_subplot(2,2,4)
plt.plot(X, Y4)
plt.show()

```
