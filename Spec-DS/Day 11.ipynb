Simple Line Plots	
    Adjusting the Plot: Line Colors and Styles 
	Adjusting the Plot: Axes Limits
	Labeling Plots

```python
#Linspace
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0 ,5, 100)
#ax.plot(x,np.sin(x))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

```

```python
#Adjusting the Plot: Line Colors and Styles

 #Line Colors
#By name (green, red,blue )
#by hex code eg: #000000 ->black #ffffff ->white  # refer https://www.color-hex.com/color/ffffff
#Gray scale code (0.75) 0 to 1
# by short rgb (rgbcmyk) (g, r, b, y ,v)
#plt.plot(x, np.sin(x), color='green')

# Line styles : dashed, solid, dotted, dashdot
# '-' =>solid
# '--' =>dashed
# '-.' =>dashdot
# ':' =>Dotted
# '' =>None
# 'o'
plt.plot(x, np.sin(x), color='#000000', linestyle='dotted')

```

```python
#Adjusting the Plot: Axes Limits
#xlim(x,y)
#ylim(x,y)

plt.plot(x, np.sin(x), color='#000000', linestyle='solid')
plt.xlim(0,6)
plt.ylim(0,6)

#Bounding concept

plt.axis('tight')
plt.axis('equal')
```

```python
#Labeling Plots: title
#1.Title
#2.Label , Legend
#3.X, Y axis
plt.plot(x, np.sin(x), color='#000000', linestyle='solid' , label='sin(x)')
plt.plot(x, np.cos(x), color='blue', linestyle='solid' , label='cos(x)')
plt.title('Sin Black solid wave', fontsize=10,color='Green')
plt.xlabel('x')
plt.ylabel('Sin & Cos Wave')
plt.legend()

```

```python
ax=plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlabel='x axis', ylabel='Y axis', title='Consolidated',xlim=(0,20),ylim=(0,20))
```

Visualizing Errors	
    1.Basic Errorbars
	2.Continuous Errors

```python
#Basic Errorbars
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,50)
a=0.8
y=np.sin(x)+a*np.random.randn(50)
plt.errorbar(x,y,yerr=a,fmt='.k')
#plt.errorbar(x,y,yerr=a,fmt='o', ecolor='lightgray', elinewidth=3) #Styles to error bar
# fmt =>format code (.k ->line , 'o'->dot)
```

```python
#Continuous Errors=>Gaussian , sklean lib
# fit(), dfit, xfit, y fit , plt.fill_between
from sklearn.gaussian_process import GaussianProcessRegressor
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 50)
model=lambda x:x*np.sin(x)
xdata=np.array([1,3,5,6,8])
ydata=model(xdata)
print(xdata)
print(ydata)
gp=GaussianProcessRegressor()
gp.fit(xdata[:,np.newaxis], ydata)
xfit =np.linspace(0,10,1000)
yfit, dyfit_ori=gp.predict(xfit[:,np.newaxis], return_std=True)
dyfit=2*dyfit_ori
plt.plot(xdata, ydata, 'o')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit-dyfit, yfit+dyfit, color='gray')
```

Density and Contour Plots	
    Visualizing a Three-Dimensional Function
	Histograms, Binnings, and Density
	Two-Dimensional Histograms and Binnings

```python
#Visualizing a Three-Dimensional Function
import matplotlib.pyplot as plt
import numpy as np
def f(x,y):
    return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)

x=np.linspace(0,5,50)
y=np.linspace(0,5,40)
X, Y =np.meshgrid(x,y)
Z=f(X,Y)
#plt.contour(X,Y,Z)
#plt.contour(X,Y,Z,20, cmap='RdGy') #->20 is interval space, RdGy->RedGray
#plt.contourf(X, Y, Z, 20, cmap='RdGy')
con=plt.contourf(X, Y, Z, 3, color='Black')
plt.clabel(con, inline=True,fontsize=7)
```

```python
#Histograms, Binnings, and Density
import numpy as np
import matplotlib.pyplot as plt
a=np.random.randn(1000)
#plt.hist(a)
plt.hist(a,histtype='stepfilled',color='green',bins=30)
```

```python
a=np.random.normal(0,0.8,1000)
b=np.random.normal(-2, 1,1000)
c=np.random.normal(3,2,1000)
plt.hist(a,histtype='stepfilled',bins=30)
plt.hist(b,histtype='stepfilled',bins=30)
plt.hist(c,histtype='stepfilled',bins=30)
```

```python
#Two-dimensional histogram - plt.hist2d

plt.hist2d(a,b,bins=30, cmap='Blues')
```

```python
#Hexagonal binnings
plt.hexbin(a,b,gridsize=30, cmap='Blues')
```

```python
#Kernel density estimation

from scipy.stats import gaussian_kde
data=np.vstack([a,b])
kde=gaussian_kde(data)
xgrid=np.linspace(-3.5,3.5,40)
ygrid=np.linspace(-6,6,40)
Xgrid,Ygrid=np.meshgrid(xgrid, ygrid)

Z=kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
plt.imshow(Z.reshape(Xgrid.shape), cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```
